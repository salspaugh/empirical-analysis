## Reading in the data

```python
def read_keystrokes(keystroke_filename):
    keystrokes = []
    with open(keystroke_filename) as keystroke_file:
        for line in keystroke_file.readlines():
            record = line.strip().split()
            timestamp = float(record[0])
            user = record[1]
            keystrokes.append((timestamp, user))
    return keystrokes
```

```python
keystrokes = read_keystrokes("keystroke-timing.txt")
```

### 3. There's a large body of theoretical modeling premised on the notion that the "arrival" of new users (i.e., the times at when they first show up) looks like a Poisson process. To the degree that you can, do you believe that this data supports that premise? Explain how you arrived at your conclusion. Recall that Poisson processes have exponentially-distributed interarrivals, i.e., if you compute the time difference between consecutive arrivals, it should be exponentially distributed. Note that you'll have to do a bit of processing of the data to extract the interarrivals between new users.

First we'll examine the pattern of interarrival times and the distribution of interarrival times.

```python
def compute_interarrivals(seq):
    seq.sort()
    return [seq[idx+1] - seq[idx] for idx, s in enumerate(seq[:-1])]

# Plot interarrival intervals over users.
user_interarrivals = compute_interarrivals(arrival_times)
def plot_interarrivals(interarrivals):
    plt.plot(interarrivals, color='black')
    plt.xlabel("Arrival Index", fontsize=16)
    plt.ylabel("Seconds Since Last Event", fontsize=16)

plot_interarrivals(user_interarrivals)
```

```python
# Plot interarrival time histogram.
def plot_interarrival_histogram(interarrivals, ymax=65, log=False):
    plt.ylim(ymin=0, ymax=ymax) # Necessary due to bug for histtype='step' below.
    n, bins, patches = plt.hist(interarrivals, num_bins, histtype='step', color='black')
    plt.xlabel("Interarrival Time Bin", fontsize=16)
    plt.ylabel("Count", fontsize=16)
    return n, bins, patches
    
_, _, _ = plot_interarrival_histogram(user_interarrivals)
```

```python
# Plot the empirical PDF of interarrival times and 
# the exponential PDF based on the empirical average of interarrival times.
def plot_interarrival_empirical_pdf(interarrivals, ymax=1.):

    plt.ylim(ymin=0, ymax=ymax) # Necessary due to bug for histtype='step' below.
    pdf, bins, patches = plt.hist(interarrivals, num_bins, histtype='step', color='black', normed=True)
    plt.xlabel("Interarrival Time", fontsize=16)
    plt.ylabel("Empirical Probability", fontsize=16)

    lmbda = 1/np.mean(interarrivals)
    dom = range(int(np.ceil(max(interarrivals))))
    exp_pdf = [lmbda * np.exp(-lmbda*x) for x in dom]
    plt.plot(dom, exp_pdf, color='red', linewidth=2)
    return pdf, bins, patches
    
pdf, bins, _ = plot_interarrival_empirical_pdf(user_interarrivals, ymax=.15)
print "Area under the curve:", np.sum(pdf * np.diff(bins)) # Check that it is a PDF.
```

This looks like a *reasonably* close fit. Next we'll try a statistical test.

SciPy offers much fewer statistical tests than e.g., MATLAB, so we will use just one simple one as suggested on [stats.stackexchange](http://stats.stackexchange.com/a/32086).

This is a one-sided version of [Grubb's test for outliers](http://en.wikipedia.org/wiki/Grubbs%27_test_for_outliers).

I have no idea if the above application of this test is correct. I briefly assessed and rejected other options:

1. **Use the log-likelihood ratio statistic against the Chi-squared distribution:** I didn't know what the null distribution should be for this statistic, otherwise I would have used it. This was suggested [here](http://stats.stackexchange.com/a/1212) along with two other suggestions that I don't know how to apply.

2. **Use Lilliefors test for exponentiality:** This is implemented in MATLAB but not SciPy. It requires the Lilliefors distribution which [according to Wikipedia](http://en.wikipedia.org/wiki/Lilliefors_test) has only been computed by MCMC methods to date, so I'm not about to implement that.

3. **Moran or Bartlett's test:** [This test](http://en.wikipedia.org/wiki/Bartlett's_test) was also suggested on [stats.stackexchange](http://stats.stackexchange.com/a/32084) looks somewhat straightforward but I am just not sure it's testing the right thing eitehr so I'm only going to try the one.

```python

def is_exponential_based_on_grubbs(interarrivals):
    import scipy.stats
    Ymin = min(interarrivals)
    Ybar = np.mean(interarrivals)
    s = np.std(interarrivals, ddof=1) # Use unbiased sample standard deviation -- is this right?
    grubbs = (Ybar - Ymin) / s
    print "Grubbs's test statistic: %.5f" % grubbs
    alpha = .05 # significance
    N = len(interarrivals)
    upper_critical_value = scipy.stats.t.pdf(alpha/N, N-2)
    test = ((N-1)/np.sqrt(N)) * np.sqrt(upper_critical_value**2 / (N - 2 + upper_critical_value**2))
    print "Test %.5f:" % test
    is_exponential = grubbs > test # It doesn't follow a normal distribution, it follows an exponential one.
    print "Follows an exponential distribution: %s" % str(is_exponential)
    return is_exponential

_ = is_exponential_based_on_grubbs(user_interarrivals)
```

This statistic in combination with the above graph suggests that the user arrival time is indeed Poisson.

```python
import scipy.stats

def plot_cdf(seq):
    hist, l, b, e = scipy.stats.cumfreq(seq, 50)
    plt.plot(np.linspace(l, l+(b*50), 50), hist/len(seq), color='black')
    plt.xlim(xmin=0)

plot_cdf(pooled_interarrivals)
```

This looks terrible too, all you can see are the outliers. Let's try to remove those. We will use [Tukey and Hoaglin's definition of quartiles](http://mathworld.wolfram.com/Quartile.html) and remove all [points that lie 1.5 times the interquartile range beyond the first or third quartile](http://mathworld.wolfram.com/Outlier.html).

```python
n = len(pooled_interarrivals)
k = 2 if n % 2 == 0 else 1
first_quartile_idx = (n + k + 1)/4
first_quartile = pooled_interarrivals[first_quartile_idx]
third_quartile_idx = (3*n + k)/4
third_quartile = pooled_interarrivals[third_quartile_idx]
iqr = third_quartile - first_quartile

low = pooled_interarrivals < first_quartile - 1.5*iqr
high = pooled_interarrivals > third_quartile + 1.5*iqr
low_outliers = np.extract(low, pooled_interarrivals)
high_outliers = np.extract(high, pooled_interarrivals)
neither = np.logical_and(np.logical_not(low), np.logical_not(high))
not_outliers = np.extract(neither, pooled_interarrivals)

print "First quartile index: %d" % first_quartile_idx
print "First quartile value: %f" % first_quartile
print "Third quartile index: %d" % third_quartile_idx
print "Third quartile value: %f" % third_quartile
print "Interquartile range: %f" % iqr 
print "Number of high outliers: %d" % len(high_outliers)
print "Number of low outliers: %d" % len(low_outliers)
print "Total outliers: %d" % (len(low_outliers) + len(high_outliers))
print "Percent outliers: %.2f" % ((len(low_outliers) + len(high_outliers))*1./n*100.)
```

### 5. The aggregate set of interarrivals that you extracted has more than one qualitatively different type of behavior. See to what degree you can identify these, and explain how you did so.

```python
def print_pct_in_interval(seq, low=None, high=None):
    n = len(seq)
    if low is None and high is None:
        return 100. 
    if low is not None:
        high_enough = seq > float(low)
        condition = high_enough
    if high is not None:
        low_enough = seq <= float(high)
        condition = low_enough
    if low is not None and high is not None:
        condition = np.logical_and(low_enough, high_enough)
    interval = np.extract(condition, seq)
    pct = len(interval)*1./n*100
    print "Percent in interval: %.2f" % pct
    
print_pct_in_interval(pooled_interarrivals, high=.01)
print_pct_in_interval(pooled_interarrivals, high=.1)
print_pct_in_interval(pooled_interarrivals, high=.25)
print_pct_in_interval(pooled_interarrivals, high=.5)
print_pct_in_interval(pooled_interarrivals, high=.75)
print_pct_in_interval(pooled_interarrivals, high=1)
print_pct_in_interval(pooled_interarrivals, high=10)
print_pct_in_interval(pooled_interarrivals, high=60)
print_pct_in_interval(pooled_interarrivals, high=300)
print_pct_in_interval(pooled_interarrivals, low=.1, high=.5) # Normal typing rate?
print_pct_in_interval(not_outliers, low=.1, high=.5)

def plot_log_cdf(seq):
    seq = np.log(seq)
    plot_cdf(seq)

plot_log_cdf(pooled_interarrivals)
```
