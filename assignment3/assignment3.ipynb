First import the packages we'll need and define some constants.

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
import os
import os.path
import random
import scipy.stats
%matplotlib inline

TWEET_TIMINGS = "tweet-timings"
EPSILON = .001
```

Read in all of the user timing data.

```python
def read_timing(f):
    with open(f) as f:
        return [float(line.strip()) for line in f.readlines()]

timing_files = [ f for f in os.listdir(TWEET_TIMINGS) if os.path.isfile(os.path.join(TWEET_TIMINGS, f)) ]
user_timings = { user: read_timing(os.path.join(TWEET_TIMINGS, user)) for user in timing_files }
```

How many users are there?

```python
nusers = len(user_timings)
print "Number of users: %d" % nusers
```

How many tweets does each user have?

```python
ntweets = [ len(timings) for timings in user_timings.values() ]
ntweets.sort(reverse=True)
rank = range(nusers)
plt.bar(rank, ntweets)
plt.xlim(xmax=nusers)
plt.xlabel("Rank", fontsize=16)
_ = plt.ylabel("Number of Tweets", fontsize=16)
```

That's very strange! What's with those three plateaus? And that spike at the beginning is strange. At first I thought maybe these plateaud tweet numbers could be bots, but then I checked the user names that have one of those values and some of them are verified accounts (e.g., randizuckerberg).

I want to look at this distribution of number of tweets per users another way to see what the median and spread look like, so I'll try a boxplot.

```python
plt.boxplot(ntweets)
plt.ylabel("Number of Tweets", fontsize=16)
print "Min number of tweets: %d" % min(ntweets)
print "Max number of tweets: %d" % max(ntweets)
print "Median number of tweets: %d" % np.median(ntweets)
```

This is consistent with the distribution of number of tweets per user.

Next I want to know how many tweets there are in total, and over what period of time these tweets were issued. I also want to know roughly how many tweets per day there are here.

```python
timings = list(itertools.chain.from_iterable(user_timings.values()))
timings.sort()
earliest_time = timings[0]
print "Earliest time: %d" % earliest_time
duration_secs = timings[-1] - earliest_time
duration_days = duration_secs/60./60./24.
print "Total timings: %f" % len(timings)
print "Duration (seconds): %f" % duration_secs
print "Duration (days): %f" % duration_days
print "Duration (years): %f" % (duration_days/365.)
print "Max tweets per day: %f" % (max(ntweets)/duration_days)
```

There are about 134K tweets represented, over a period of about two years, starting on Wed, 16 Apr 2008 22:17:51 GMT. This amounts to about two to three tweets per day, but some users may have a higher rate. 

Since things were sampled weirdly, I'm skeptical that the duration is meaningful, or the number of tweets really (i.e., these probably don't represent the all tweets in a given time period).

Now I want to look at the duration of timestamps per user -- the earliest and last tweet seen by each user.

```python
user_durations = {}
for (user, timings) in user_timings.iteritems():
    timings.sort()
    user_durations[user] = timings[-1] - timings[0]
durations = sorted(user_durations.values(), reverse=True)
plt.bar(rank, durations)
plt.xlim(xmax=nusers)
plt.xlabel("Rank", fontsize=16)
_ = plt.ylabel("Duration of Timings", fontsize=16)
```

I don't know what to conclude from this except to say that this doesn't look too suspicious to me. It doesn't mean that much to me aside from that.

Now I want to know how the duration of a user's session relates to the number of tweets in that session. I imagine that high velocity, low-duration tweets are bots. I expect to see weird bands due to the strange sampling plateaus.

```python
duration_ntweets = [(user_durations[user]/60./60./24., len(user_timings[user])) for user in user_timings]
x_durations, y_ntweets = zip(*duration_ntweets)
plt.scatter(x_durations, y_ntweets)
plt.xlim(xmin=0, xmax=max(x_durations)+50)
plt.xlabel("Timing Duration (Days)", fontsize=16)
plt.ylim(ymin=0, ymax=max(y_ntweets)+100)
_ = plt.ylabel("Number of Total Tweets", fontsize=16)
```

The strange bands are indeed there. There are only a few very high-velocity but low-duration tweets. The duration of the bands seem spread out, so maybe what was done is they just took N consecutive tweets from users.

Now I want to know what the number of tweets per day looks like. We saw the average is about two to three, but I bet the actual rate is more spiky.

```python
def increment_timing_bucket(bins, time):
    bins[int(time)] += 1
    return bins

def bin_tweets_per_day(timings):
    timings.sort()
    duration_days = (timings[-1] - timings[0])/60./60./24.
    timings = [(t - timings[0])/60./60./24. for t in timings]
    bins = range(int(duration_days)+1)
    ntimings = { b:0 for b in bins }
    reduce(increment_timing_bucket, timings, ntimings)
    frequencies = [count for (bucket, count) in sorted(ntimings.items())]
    return bins, frequencies
    
def tweets_per_day(timings):
    bins, frequencies = bin_tweets_per_day(timings)
    plt.bar(bins, frequencies)
    plt.xlim(xmax=bins[-1])
    plt.xlabel("Day", fontsize=16)
    plt.ylabel("Number of Tweets", fontsize=16)
    
timings = list(itertools.chain.from_iterable(user_timings.values()))
tweets_per_day([t-earliest_time for t in timings])
```

This looks bad. There are very few tweets for about a year and a half, then there is a huge spike in tweets, followed by another small spike in tweets. Again due to the weird sampling, this is really hard to interpret. 

I want to look at this in log scale so that I can see if there were zero tweets at the beginning or if that's just a scaling issue.

```python
plt.bar(bins, [max(np.log10(f+EPSILON)+1., 0.) for f in frequencies])
plt.xlim(xmax=bins[-1])
plt.xlabel("Day", fontsize=16)
plt.yticks(range(0, 6), range(-1,5))
_ = plt.ylabel("Log Number of Tweets", fontsize=16)
```

So it looks like in fact there were close to zero tweets per day for almost a year, and then only a few per day for another half year or so. This data set is frustrating.

Now I want to see if the tweets occur according to some pattern, like every minute, or every five minutes, or every hour, because we're looking for bots, and I would expect that bots tweet at that sort of interval.

```python
modulo = 60

timings = list(itertools.chain.from_iterable(user_timings.values()))

def increment_mod_bucket(bins, time):
    bins[time] += 1
    return bins

mod_timings = [t%modulo for t in timings]
bins = range(modulo)
ntimings = { b:0 for b in bins }
reduce(increment_mod_bucket, mod_timings, ntimings)
frequencies = [count for (bucket, count) in sorted(ntimings.items())]
plt.bar(bins, frequencies)
plt.xlim(xmax=modulo)
plt.xlabel("Time Modulo %d" % modulo, fontsize=16)
plt.ylabel("Number of Tweets", fontsize=16)
print "Skew of times %f" % scipy.stats.skew(frequencies)
```

Sadly I don't see any suspicious patterns jumping out here. I tried various modulos and that didn't help. I computed skew of this distribution so that I could compare it to the skew below, since this distribution looks pretty uniform to me. More on that later.

Next I want to start looking at interarrivals, because I'm not having much luck looking at arrivals. (Or perhaps there aren't many bots in this data set.) First I will look at the distribution of interarrival times, where interarrivals are computed per-user and then aggregated. I'm doing it this way because we were instructed to do it this way in the first homework and I guess it kind of makes sense, though again because of the unknown and mysterious sampling, maybe these are also not meaningful.

```python
timings = list(itertools.chain.from_iterable(user_timings.values()))

def get_interarrivals(times):
    times.sort()
    interarrivals = []
    for idx, curr in enumerate(times[:-1]):
        next = times[idx+1]
        interarrivals.append(next - curr)
    return interarrivals

user_interarrivals = { user: get_interarrivals(timings) for (user, timings) in user_timings.iteritems() }
all_interarrivals = list(itertools.chain.from_iterable(user_interarrivals.values()))
log_all_interarrivals = [np.log10(t) for t in all_interarrivals]
plt.boxplot(log_all_interarrivals)
plt.ylabel("Log Interarrival Time", fontsize=16)
print "Min interarrival time: %d" % min(all_interarrivals)
print "Max interarrival time: %d" % max(all_interarrivals)
print "Median interarrival time: %d" % np.median(all_interarrivals)
```

The shortest time between tweets is zero, while the longest is about six months. The median is closer to five minutes. That seems shorter than I would have expected, based on my own Twitter usage.

Next I'll look at the distribution of interarrivals. 

```python
all_interarrivals = list(itertools.chain.from_iterable(user_interarrivals.values()))
all_interarrivals = [t for t in all_interarrivals if t < 720]
bins = range(int(max(all_interarrivals))+1)
ninterarrivals = { b:0 for b in bins }
reduce(increment_timing_bucket, all_interarrivals, ninterarrivals)
frequencies = [count for (bucket, count) in sorted(ninterarrivals.items())]
plt.bar(bins, frequencies)
plt.xlim(xmax=bins[-1])
plt.xlabel("d", fontsize=16)
plt.ylabel("Num Interarrivals == d", fontsize=16)

timings = list(itertools.chain.from_iterable(user_timings.values()))
rate = len(timings)/duration_secs
print "Number of tweets: %d" % len(timings)
print "Tweets per second: %f" % rate

exponential = [rate*np.exp(-rate*x) for x in bins]
_ = plt.plot(bins, exponential, color="red", linewidth=5)

rates = [len(t)/user_durations[user] for (user, t) in user_timings.iteritems() if len(t) > 0]
rate = np.mean(rates)
print "Average tweets per second per user: %f" % rate
print "Median tweets per second per user: %f" % np.median(rates)

exponential = [rate*np.exp(-rate*x) for x in bins]
_ = plt.plot(bins, exponential, color="blue", linewidth=5)
```

I had to look at only interarrivals that were less than 12 minutes apart because otherwise the domain was so large that you can't see the obvious adn suspicious spikes at around zero or one, 60, 300, and 600. The spikes are probably due to bots! It doesn't tell us how many bots there are though.

I tried fitting an exponential to these but it didn't work. I'm either doing it wrong, or I need to try to fit some other curve to it. 

Since I can't make that work, I'm going to forget about it and look at the arrrival rate per user, because some of these one-second intervals make me think that some users must have really high tweet rates.

```python
rates.sort(reverse=True)
ranks = range(len(rates))
plt.bar(ranks, rates)
plt.xlim(xmax=ranks[-1])
plt.xlabel("User rank", fontsize=16)
_ = plt.ylabel("Arrival rate", fontsize=16)
```

There definitely looks to be some suspiciously high tweet rates. There seems to be a bit of an inflection at about 30 users in. Maybe the first 30 users are bots! This would mean that about 5% of users are bots.

I want to go back to examining those interarrival spikes. I'll look at the interrival times modulo 60 so I can see if there are spikes near zero, where I would expect them.

```python
all_interarrivals = list(itertools.chain.from_iterable(user_interarrivals.values()))
modulo = 60
mod_interarrivals = [t%modulo for t in all_interarrivals]
bins = range(modulo)
ninterarrivals = { b:0 for b in bins }
reduce(increment_mod_bucket, mod_interarrivals, ninterarrivals)
frequencies = [count for (bucket, count) in sorted(ninterarrivals.items())]
plt.bar(bins, frequencies)
plt.xlim(xmax=modulo)
plt.xlabel("Interarrival Time Modulo %d" % modulo, fontsize=16)
_ = plt.ylabel("Number of Interarrivals", fontsize=16)
```

Indeed, there is a lot of skew in interarrival times around zero, meaning there are a bunch of tweets coming in at intervals that are some multiple of 60 seconds. It is suprising to see a lot of skew here and not in the arrival times. So at least bots are smart enough not to tweet on the minute exactly?

I want to try to quantify the above behavior so that I can get a number for each user that describes how suspicious their interarrivals are. I think that this distribution above looks almost uniform, but with a big spike at one end, causing it to eb very skewed. So I'll compute skewness for the interarrivals for each user and examine those. First I want to see the skewness of the above distribution and compare that to less skewed disributions, so that I know what kind of numbers we're dealing with.

```python
print "Skew of interarrivals %f" % scipy.stats.skew(frequencies)
for noise in [10**r for r in range(5)]:
    print "Skew of uniform-ish at 2000 %f" % scipy.stats.skew([2000+noise*random.random()*random.randint(-1,1) for f in frequencies])
```

The aggregated per-user interarrival distribution is very skewed. More uniform distributions have skews closer to zero.

Given this information, I think that users who both have very high tweet rates and a big skew in their interarrival distribution will probably be bots. I now will try to estimate how many users this might be by plotting interarrival skew versus tweet rate for each user, and counting how many fall into a certain region of the plot.

I make two estimates, a conservative one that allows for more skew and a higher tweet rate (based on an estimate of the tweet rate of a real human I know who tweets seemingly incessantly), and less conservative estimate that is based on how often I tweet and a more uniform distribution of interarrival times.

```python
skew_rate = []
for user in user_interarrivals:
    mod_interarrivals = [t%60 for t in user_interarrivals[user]]
    skew = scipy.stats.skew(mod_interarrivals)
    ntweets = len(user_timings[user])
    duration = user_durations[user]
    rate = ntweets/duration
    skew_rate.append((skew, rate))
skew_per_user, rate_per_user = zip(*skew_rate)
plt.scatter(skew_per_user, [np.log10(r) for r in rate_per_user])
plt.xlabel("Skew in Interarrivals Mod 60", fontsize=16)
plt.ylabel("Log Arrival Rate", fontsize=16)

cons_bound_tweets_per_sec = (48*1)/(60.*60.*24)
print "Conservative upper bound on tweets per second: %f (log10=%f)" % (cons_bound_tweets_per_sec, np.log10(cons_bound_tweets_per_sec))
lib_bound_tweets_per_sec = (2*1)/(60.*60.*24)
print "Less conservative upper bound on tweets per second: %f (log10=%f)" % (lib_bound_tweets_per_sec, np.log10(lib_bound_tweets_per_sec))

print 

bad = [1. if ((s>1.0 or s<-1.0) and (r >= cons_bound_tweets_per_sec)) else 0. for (s,r) in skew_rate]
print "Conservative estimate of number of bots: %d (%f percent)" % (sum(bad), (sum(bad)*100./nusers))
bad = [1. if ((s>.5 or s<-.5) and (r >= lib_bound_tweets_per_sec)) else 0. for (s,r) in skew_rate]
print "Less estimate of number of bots: %d (%f percent)" % (sum(bad), (sum(bad)*100./nusers))
```

In the end, I estimate that between 2.5% and 8% of the users in this data set are bots. 

The upper end of this estimate compares favorably with articles from August 2014 (a few months ago) that estimate the number of automated Twitter accounts to be 23M out of 271M total users, or 8.5%.  Because this data is from six years prior, I would not expect the percent of bots to be the same in it as in 2014, but I would expect it to be somewhat close, because bots were most likely already a thing in 2008, so I wouldn't expect a big surge in proportion later.

This estimate also compares favorably to my back-of-the-envelope estimate from before that about 30 of the users (or 5%) had such a high tweet rate that they seemed like bots.
